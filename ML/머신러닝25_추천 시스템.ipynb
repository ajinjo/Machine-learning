추천 시스템

- 지금은 추천 시스템(Recommendations) 전성시대
- 전자 상거래부터 유튜브, 애플 뮤직 등 컨텐츠 포털까지 추천 시스템을 통해 
- 사용자의 취향을 이해하고 맞춤 상품과 컨텐츠를 제공해서 
- 고객을 사이트에 조금이라도 더 머무르게 하기 위해 전력을 기울이고 있음


- 발전하는 추천 엔진으로 인해 더욱더 많은 수익을 올리 수 있다는 사실을 알고
- 주요 업체들은 추천 엔진의 고도화에 큰 비용과 노력을 들이고 있고
- 실제로 많은 전자상거래 업체가 추천 시스템을 도입하여 
- 메출을 큰 폭으로 증가 시키고, 사용자의 쇼핑 즐거움 또한 증가


- 개인적 취향을 정확히 이해하고 추천
- 하나의 컨텐츠를 선택했을 때 선택된 컨텐츠와 연관된 추천 컨텐츠가 얼마나 사용자의 관심을 끌고 
- 개인에게 맞춘 컨텐츠를 추천했는지는 그 사이트이 평판을 좌우하는 매우 중요한 요소
- 추천 시스템을 접한 사용자는 해당 사이트를 더 강하게 신뢰하게 되어 더 많은 추천 컨텐츠 선택하게 되고
- 결국 더 많은 데이터가 추천 시스템에 축적되면서 추천이 더욱 정확해지고 다양한 결과를 얻을 수 있는
- 좋은 선순환 시스템 구축 가능
- 치열한 경쟁 환경에 놓여 있는 전자 상거래, 영화/음악 컨텐츠 제공 업체 등은 
- 사용자의 관심을 오랫동안 지속하기 위해 똑똑한 추천 시스템을 갖춰야 함을잘 알고 있기에
- 추천 시스템의 고도화에 큰 비용을 들이고 투자를 아끼지 않음

온라인 스토어의 필수 요소, 추천 시스템
- 온라인 컨텐츠 제공 업체의 다양한 상품 이미지와 복잡한 카테고리, 메뉴 구성 등으로
- 온라인상 제품 선택의 어려움 발생 -> 부정적인 이미지 -> 매출 감소
- 좋은 추천 시스템 : 사용자가 원하는 것을 빠를게 찾아내 사용자의 온라인 쇼핑 편리함 증대
- 온라인 스토어에서 추천 시스템은 필수 구성 요소

추천 시스템을 구성하는데 사용 가능한 데이터
- 사용자가 어떤 상품을 구매했는가?
- 사용자가 어떤 상품을 둘러보거나 장바구니에 넣었는가?
- 사용자가 평가한 영화 평점은?
- 제품 평가는?
- 사용자가 스스로 작성한 자신의 취향은?
- 사용자가 무엇을 클릭했는가?

이러한 데이터를 기반으로 추천 시스템이 사용자가 상품을 구매하도록 유혹하는 문구
- 당신만을 위한 최신 상품
- 이 상품을 선택한 다른 사람들이 좋아하는 상품들
- 이 상품을 좋아하나요?

추천 시스템의 유형
- 컨텐츠 기반 필터링(Content based filtering)
- 협업 필터링(Collaborative Filtering)  
  - 최근접 이웃 협업 필터링 (Nearest Neighbor)    
  - 잠재 요인 협업 필터링 (Latent Factor)


- 추천 시스템의 초창기에는 컨턴츠 기반 필터링이나 최근접 이웃 기반 협업 필터링 주로 사용
- 넷플릭스 추천 시스템 경연 대회에서 
- 행렬 분해 (Matrix Factorization) 기법을 이용한 잠재 요인 협업 필터링 방식이 우승하면서
- 대부분의 온라인 스토어에서 잠재 요인 협업 필터링 기반의 추천 시스템 적용
- 그러나 서비스하는 아이템의 특성에 따라
- 컨턴체 기반 필터링이나 최근접 이웃 기반 협업 필터링 방식을 유지하는 사이트도 많고
- 아마존의 경우는 아직도 아이템 기반의 최근접 이웃 협업 필터링 방식을 추천 엔진으로 사용
- 요즘에는 개인화 특성을 좀 더 강화하기 위해서
- 하이브리드 형식으로 컨텐츠 기반과 협업 기반을 적절히 결합해 사용하는 경우도 증가
